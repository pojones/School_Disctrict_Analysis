```python
prefixes_suffixes = ["Dr. ", "Mr. ","Ms. ", "Mrs. ", "Miss ", " MD", " DDS", " DVM", " PhD"]
# start a list and fill it with prefixes and suffixes for removal

for word in prefixes_suffixes:
# cycle through list 'prefixes_suffixes'
    student_data_df["student_name"] = student_data_df["student_name"].str.replace(word,"")
    # in dataframe 'stude...df', convert 'student name' object to string and replace every word with empty string

student_data_df.head(10)
# print the first ten rows of dataframe to see if our prefix/suffix replacement worked
```

```python
passing_reading_df = school_data_complete_df[school_data_complete_df["reading_score"] >= 70]
# gather all the students that are passing reading and put them in dataframe 'passing_reading_df'
passing_reading_df.head()
# print first five entries of dataframe 'passing_math'
```

```python
passing_math_count = passing_math_df["student_name"].count()
# count the number of students in 'passing_math' 

passing_reading_count = passing_reading_df["student_name"].count()
# count the number of students in 'passing_reading' 

print(passing_math_count)
print(passing_reading_count)
```

```python
passing_math_percentage = passing_math_count / float(student_count) * 100
# calculate percentage of students that passed math

passing_reading_percentage = passing_reading_count / float(student_count) * 100
# calculate percentage of students that passed reading

print(passing_math_percentage, passing_reading_percentage)
```

```python
# Calculate the students who passed both math and reading.
passing_math_reading = school_data_complete_df[(school_data_complete_df["math_score"] >= 70) & (school_data_complete_df["reading_score"] >= 70)]
# from the complete dataframe, we select students who have passed math and reading and placed them in new dataframe
passing_math_reading.head()
# print the dataframe's first five lines
```

```python
overall_passing_math_reading_count = passing_math_reading["student_name"].count()
# count the number of students in passing_math_reading list
overall_passing_math_reading_count
# print the number of students who passed both math and reading
```

```python
overall_passing_percentage = overall_passing_math_reading_count / student_count * 100
# calculate the ratio of students who passed both math and reading to the total student count
overall_passing_percentage
# print the percentage number of students who passed both math and reading
```

```python
district_summary_df = pd.DataFrame(
          [{"Total Schools": school_count,
          "Total Students": student_count,
          "Total Budget": total_budget,
          "Average Math Score": average_math_score,
          "Average Reading Score": average_reading_score,
          "% Passing Math": passing_math_percentage,
         "% Passing Reading": passing_reading_percentage,
        "% Overall Passing": overall_passing_percentage}])
        # creates a new dataframe with specific keys, arranged with a list of values
district_summary_df
# print the dataframe
```

```python
# Define a function that calculates the percentage of students that passed both 
# math and reading and returns the passing percentage when the function is called.

def passing_math_percent(pass_math_count, student_count):
# define function that takes two variables 'pass_math_count' and 'student_count'
    return pass_math_count / float(student_count) * 100
    # return percentage of students that passed math
```

```python
passing_math_percent(passing_math_count, student_count)
# call the 'passing_math_percent' function, using our student count variables from before
```

```python
district_summary_df["Total Students"] = district_summary_df["Total Students"].map("{:,}".format)
# format the 'Total Students' section of the district_summary_df to have a comma for the thousands separator

district_summary_df["Total Students"]
# print the new 'Total Students' section from the dataframe
```

```python
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("${:,.2f}".format)
# format the 'Total Budget' section of the dataframe to have comma for thousands separator, decimal separator, and a "$".

district_summary_df["Total Budget"]
# print the 'Total Budget' section of dataframe
```

```python
district_summary_df["Average Math Score"] = district_summary_df["Average Math Score"].map("{:.1f}".format)
# format 'Average Math Score' to have a single decimal spot
district_summary_df["Average Reading Score"] = district_summary_df["Average Reading Score"].map("{:.1f}".format)
# format 'Average Reading Score' to have a single decimal spot
district_summary_df["% Passing Math"] = district_summary_df["% Passing Math"].map("{:.0f}".format)
# format '% Passing Math' to have no decimal
district_summary_df["% Passing Reading"] = district_summary_df["% Passing Reading"].map("{:.0f}".format)
# format '% Passing Reading' to have no decimal 
district_summary_df["% Overall Passing"] = district_summary_df["% Overall Passing"].map("{:.0f}".format)
# format '% Overall Passing' to have no decimal
```

```python
district_summary_df
# display district_summary_df
```

```python
new_column_order = ["Total Schools", "Total Students", "Total Budget","Average Math Score", "Average Reading Score",
                    "% Passing Math", "% Passing Reading", "% Overall Passing"]
                    # place new columb order in a list

district_summary_df = district_summary_df[new_column_order]
# assign dataframe new column order
district_summary_df
# display the dataframe
```
